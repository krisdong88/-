Ferret UI 2024年4月8号

主要内容和贡献
模型介绍：介绍了一个名为 Ferret-UI 的新型多模态大型语言模型，专门用于移动用户界面的理解。这个模型具有指向、定位和推理的能力，可以处理具体的用户界面元素并执行开放式语言指令。
技术细节：为了适应移动设备屏幕的长宽比和小型对象（如图标和文本），Ferret-UI在Ferret模型的基础上进行了改进，通过将屏幕分为子图像并分别编码，以增强细节处理能力。
数据和训练：研究团队收集了涵盖基本的语义和空间任务到高级推理任务的数据，以此训练模型。这包括识别图标、文本查找和部件列表等任务。

结论与展望
模型表现：Ferret-UI在各项基础和高级用户界面任务中表现出色，超过了多数开源的UI多模态语言模型和GPT-4V。
未来工作：文中提到了进一步提升模型性能的可能方向，包括改进数据集的覆盖度和多样性，以及继续优化模型结构以更好地处理复杂的用户界面场景。
