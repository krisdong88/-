首先机器学习分为有监督学习，无监督学习，深度学习，自监督学习，强化学习，半监督学习。然后在无监督学习中有分成很多类似，聚类，生成式学习和对比式学习，这篇笔记主要会聚焦在对比式学习，生成式学习比较著名的工作有GAN 和 VAE。对比式学习主要分成两类，第一类式同样样本来进行学习，第二类是通过不同类样本来进行学习，第一个是学习同类样本的共同之处，给上label。不同的就是学习不同之处然后区分开来。对比学习的目标是学习一个编码器，此编码器对同类数据进行相似的编码，并且使不同类的数据的编码结果尽可能的不同。

总的来说，无监督学习主要的思想就是聚类，也就是会缩小与正样本的距离，并且同时扩大与负样本的距离。
