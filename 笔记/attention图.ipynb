```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# 加载图像
image_path = 'images/zac.png'
image = Image.open(image_path)
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
# 假设attention_map是从模型中提取的attention权重，形状为 (heads, height, width)
# 这里我们简化为一个示例attention_map
attention_map = np.random.rand(8, 14, 14)  # 示例的随机权重

# 将多个heads的attention权重进行平均（如果有多个heads）
attention_map = np.mean(attention_map, axis=0)
```

```python
from scipy.ndimage import zoom

# 假设原始图像尺寸为 (H, W)
H, W = image.size
attention_map_resized = zoom(attention_map, (W/attention_map.shape[0], H/attention_map.shape[1]), order=1)
```

```python
# 将attention_map进行标准化以便可视化
attention_map_resized = (attention_map_resized - np.min(attention_map_resized)) / (np.max(attention_map_resized) - np.min(attention_map_resized))

plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.imshow(attention_map_resized, cmap='jet', alpha=0.5)  # 使用半透明的方式叠加
plt.axis('off')
plt.show()
```
