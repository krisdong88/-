Import 必要Library

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

读取图像

```python
image = cv2.imread("images\zac.png")
```

查看是否读取图片成功

```python
if image is None:
    print("Error: Could not open or find the image.")
    exit()
```

颜色转换

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

高斯模糊

```python
blurred_image = cv2.GaussianBlur(image_rgb,(9,9),0)
for i in range(1000):
    blurred_image = cv2.GaussianBlur(blurred_image, (9, 9), 0)
```

显示图像

```python
plt.subplot(121), plt.imshow(image_rgb), plt.title('Original Image')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(blurred_image), plt.title('Blurred Image')
plt.xticks([]), plt.yticks([])
plt.show()
```
