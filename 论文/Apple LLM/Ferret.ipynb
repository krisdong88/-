Ferret 2023/10/11

主要功能和创新点：
模型设计：Ferret 能够处理图像中的各种形状和粒度的空间引用，包括点、边界框和自由形状。
混合区域表示：通过结合离散坐标和连续视觉特征，Ferret 引入了一种新的混合区域表示方法，使其能够更精确地处理多样的区域输入。
空间感知视觉采样器：该模型提出了一个空间感知视觉采样器，用于从任意形状的区域提取视觉特征，解决了形状多样性带来的挑战。

数据集和评估：
GRIT 数据集：为了训练和评估 Ferret，研究者收集了名为 GRIT 的数据集，包含超过 1.1 百万个样本，这些样本涵盖了丰富的层次空间知识和硬负样本，以增强模型的鲁棒性。
评估和性能：在传统的引用和定位任务上，Ferret 显示出优于现有多模态大语言模型的性能，同时在多模态对话任务中展示了其优越的引用和定位能力。

实际应用：
开放词汇引用和定位：Ferret 不仅能理解和执行基于图像的指令，还能在对话中引用和定位图像中的具体内容，使其在实际应用中具有广泛的应用潜力，特别是在增强现实和交互式媒体等领域。

总的来说，这篇文章通过引入一种新的混合区域表示方法和空间感知视觉采样器，显著提高了模型处理图像中空间引用的能力，为深度学习和人工智能领域的研究提供了新的方向和工具。
